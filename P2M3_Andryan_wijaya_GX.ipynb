```python
# Create a data context

from great_expectations.data_context import FileDataContext

context = FileDataContext.create(project_root_dir='./')
```

```python
# Give a name to a Datasource. This name must be unique between Datasources.
datasource_name = 'Data'
datasource = context.sources.add_pandas(datasource_name)

# Give a name to a data asset
asset_name = 'Cleandata'
path_to_data = '/content/P2M3_Andryan_wijaya_data_clean.csv'
asset = datasource.add_csv_asset(asset_name, filepath_or_buffer=path_to_data)

# Build batch request
batch_request = asset.build_batch_request()
```

```python
# Creat an expectation suite
expectation_suite_name = 'expectation-Validation'
context.add_or_update_expectation_suite(expectation_suite_name)

# Create a validator using above expectation suite
validator = context.get_validator(
    batch_request = batch_request,
    expectation_suite_name = expectation_suite_name
)

# Check the validator
validator.head()
```

Expectation 1

```python
# Expectation 1 : Column `pickup_datetime` can not contain missing values

validator.expect_column_values_to_be_unique('customer_id')
```

expectation 2

```python
validator.expect_column_values_to_be_in_type_list('referred_a_friend',['bool'])
```

expectation3

```python
validator.expect_column_values_to_be_between(
    column='number_of_referrals', min_value=0, max_value=10
)
```

expectation 4

```python
validator.expect_column_values_to_be_in_set('gender', ['Male','Female'])
```

expectation 5

```python
validator.expect_column_values_to_not_be_null('country')
```

expectation 6

```python
validator.expect_column_values_to_be_between('tenure_in_months', min_value=1, max_value=73)
```

expectation 7

```python
validator.expect_column_distinct_values_to_equal_set('country',['United States'])
```

save


```python
validator.save_expectation_suite(discard_failed_expectations=False)
```

# check point

```python
# Create a checkpoint

checkpoint_1 = context.add_or_update_checkpoint(
    name = 'checkpoint_1',
    validator = validator,
)
```

```python
# Run a checkpoint

checkpoint_result = checkpoint_1.run()
```
